## Openface CSV - Gaze eye estimation

```python
import pandas as pd
```

```python
gaze = pd.read_csv("a.csv") 
```

```python
gaze.head()
```

```python
face_id =  gaze.face_id.unique()
print(face_id)
print(type(face_id[0]))
```

```python
#Unique face time stamp
for i in range(len(face_id)):
  count = 0
  print(str(face_id[i]))
  count = gaze['frame'][((gaze['gaze_angle_x']>0) & (gaze['gaze_angle_y']>0)) & (gaze['face_id'] == (face_id[i]))].count()
  print("Gaze time for face_id "+ str(i) + " " +str(count*0.04))



total_count = gaze['frame'][(gaze['gaze_angle_x']>0) & (gaze['gaze_angle_y']>0)].count()
print('total gaze time',total_count*0.04)
```
